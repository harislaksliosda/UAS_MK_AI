## Activation 

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-2,2, 100)
```

### Identity 

```python
def identity(x):
    y = x
    return y
```

```python
y = identity(x)
plt.plot(x, y)
plt.grid()
```

### Binary Step

```python

def binary_step(x):
    y=[]
    for i in range(len(x)):
        if(x[i]<0):
            y_ = 0
        elif (x[i]>=0):
            y_ = 1
        y.append(y_)
    return y
```

```python
y = binary_step(x)
plt.plot(x,y)
plt.grid()
```

### Sigmoid 

```python
def sigmoid(x):
    y = 1/(1+np.exp(-x))
    return y
```

```python
y = sigmoid(x)
plt.plot(x,y)
plt.grid()
```

### TanH

```python
y = np.tanh(x)
plt.plot(x,y)
plt.grid()
```

### arcTan

```python
y = np.arctan(x)
plt.plot(x,y)
plt.grid()
```

```python
def relu(x):
    y=[]
    for i in range(len(x)):
        if(x[i]<0):
            y_ = 0
        elif (x[i]>=0):
            y_ = x[i]
        y.append(y_)
    return y
```

```python
#rectified linear unit
y = relu(x)
plt.plot(x,y)
plt.grid()
```

### Softmax 

```python
def softmax(x, t = 1.0):
    e = np.exp(np.array(x) / t)
    y = e / np.sum(e)
    return y
```

```python
y = softmax(x)
plt.plot(x,y)
plt.grid()
```
