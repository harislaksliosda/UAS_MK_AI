```python
# Update dictionary
dict = {'Name': 'Zara', 'Age': 7, 'Class': 'First'}
dict['Age'] = 8 # update existing entry
dict['School'] = "DPS School" # Add new entry

print(dict)
print ("dict['Age']: ", dict['Age'])
print ("dict['School']: ", dict['School'])
```

```python
del dict['Name']; # remove entry with key 'Name'
dict.clear();     # remove all entries in dict
del dict ;        # delete entire dictionary
```

```python
def respon_unit(x1, x2, w1, w2, b):
    y_in = b + x1 * w1 + x2 * w2
    return y_in
respon_unit(0.5, 0.3, 0.5, -1.5, 0.7)
```

```python
x1 = 2.0
x2 = 1.5
b = 0.5
w1 = 1.1
w2 = -1.0
respon_unit(x1, x2, w1, w2, b)
```

```python
import math
def sigmoid(y_in):
    return 1 / (1+math.exp(-y_in))
```

```python
y_in = respon_unit(x1, x2, w1, w2, b)
```

```python
sigmoid(y_in)
```

## Error Handler

```python
while True:
    try:
        x = int(input("Please enter an integer number: "))
        break
    except ValueError:
        print("Oops!  That was no valid number.  Try again...")
```

```python
try: 
    nilai = int(input("Nilai: ")) # Harus Integer

except ValueError:
    print("nilai tidak sesuai")
```

```python
import sys

randomList = ['a', 0, 2]

for data in randomList:
    try:
        print("data", data)
        r = 1/int(data)
        break
    except:
        print("error!",sys.exc_info()[0],)
        print("Next .")
        print()
print("Hasil proses dari",data,"adalah",r)
```
